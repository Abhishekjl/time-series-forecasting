#### lets first make a naive model (moving average) 

```python
f_sales_base  = y.rolling(window = 4).mean() # here we can select a winndow size for moving average 
f_sales_base.plot()
```

```python
f_sales_test = pd.DataFrame({'Actual':y,'Predicted':f_sales_base})
f_sales_test.dropna(inplace = True)
```

```python
f_sales_test.plot()
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(f_sales_test['Actual'],f_sales_test['Predicted']))
```

```python
# here we can create a loop to find the best window size 
for i in range(2,15):
    try:
        f_sales_base  = y.rolling(window = i).mean() # here we can select a winndow size for moving average 
        f_sales_test = pd.DataFrame({'Actual':y,'Predicted':f_sales_base})
        f_sales_test.dropna(inplace = True)
        error = np.sqrt(mean_squared_error(f_sales_test['Actual'],f_sales_test['Predicted']))
        print('error {} and window_size {}'.format(error, i))
    except:
        pass
        
```

```python
# but the problem of moving  average is it cant be used for longer forcast and it doesnt take trend seasonality in consideration
```

### ARIMA model (combination of moving average, integration , autoregression)

ARIMA(p,d,q)
* p > 'Auto regressor' if only p given auto regression model
* d > 'integrated part'
* q > 'moving average' if only q given moving average model  


we calculate p,d,q using pacf and acf
* pacf for p 
* acf for moving average ( q) 

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plot_acf(y)
```

```python
# p = 1, q = 1, d = 0,1,2
```

```python
plot_pacf(y)
```

```python
from statsmodels.tsa.arima_model import ARIMA

import warnings
warnings.filterwarnings('ignore')
```

```python
train = y[:26]
test  = y[26:]
```

```python
arima_model = ARIMA(train,order=(4,0,4))
arima_model_fit = arima_model.fit()
```

```python
arima_model_fit.aic
```

```python
f_sales_forecast = arima_model_fit.forecast(steps = 22)[0]
```

```python
print('error',np.sqrt(mean_squared_error(test,f_sales_forecast)))
```

```python
predicted = pd.Series(f_sales_forecast, index = test.index)
```

```python
ax = y.plot(label  = 'actual')
predicted.plot(ax = ax, label = 'predicted')
plt.show()
```

### Finetuning of ARIMA model
